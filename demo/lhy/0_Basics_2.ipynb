## Edge Detection

```python
import numpy as np
import cv2 as cv
from matplotlib import pyplot as plt
```

```python
img1 = cv.imread('../../data/images/demo/p1.jpg',0)
img2 = cv.imread('../../data/images/demo/p2.png',0)
```

```python
def edge_detection(img):
    edges = cv.Canny(img,100,200)
    plt.subplot(121),plt.imshow(img,cmap = 'gray')
    plt.title('Original Image'), plt.xticks([]), plt.yticks([])
    plt.subplot(122),plt.imshow(edges,cmap = 'gray')
    plt.title('Edge Image'), plt.xticks([]), plt.yticks([])
    plt.show()
    return edges
```

```python
img1_edges = edge_detection(img1)
```

```python
img2_edges = edge_detection(img2)
```

```python
def dummy_model(edges):
    arr = edges.flatten()
    num_black_points = len(arr[arr == 0])
    print(f'The number of black points are {num_black_points}.')
    num_other_points = len(arr[arr > 0])
    print(f'The number of non-black points are {num_other_points}.')
    print(f'The ratio of black to non-black is {num_black_points / num_other_points}')
```

```python
dummy_model(img1_edges)
```

```python
dummy_model(img2_edges)
```
